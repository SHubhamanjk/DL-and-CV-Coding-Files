# **ANN Implemenatation**

```python
import pandas as pd
```

```python
data=pd.read_csv('final_data.csv')
```

```python
data
```

```python
data=data.dropna()
```

```python
data.isnull().sum().sum()
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

y= to_categorical(y)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
network = Sequential()  # Initialize a sequential neural network

network.add(Dense(units=605, activation='relu'))  # Input layer with 605 neurons and ReLU activation
network.add(Dropout(0.3))  # Dropout layer: 30% of neurons will be randomly turned off during training to prevent overfitting

network.add(Dense(units=428, activation='relu'))  # First hidden layer with 428 neurons and ReLU activation
network.add(Dropout(0.3))  # Dropout layer for regularization

network.add(Dense(units=214, activation='relu'))  # Second hidden layer with 214 neurons
network.add(Dropout(0.3))  # Dropout layer to reduce overfitting

network.add(Dense(units=107, activation='relu'))  # Third hidden layer with 107 neurons
network.add(Dropout(0.3))  # Dropout layer for further regularization

network.add(Dense(units=243, activation='softmax'))  # Output layer with 243 neurons (for multi-class classification), using softmax activation
```

```python
network.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) # Compilation of Model
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitor validation loss to decide when to stop training
    patience=5,  # Stop training if validation loss doesn't improve for 5 consecutive epochs
    restore_best_weights=True  # Restore the model weights from the epoch with the best validation loss
)  # Prevents overfitting by stopping training early when the model stops improving
```

```python
model = network.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),  # Use test data for validation to monitor performance
    epochs=100,  # Number of times the entire dataset is passed through the network during training
    batch_size=32,  # Number of samples per batch; smaller batches use less memory but may take longer to converge
    callbacks=[early_stopping],  # Apply early stopping to prevent overfitting and unnecessary training
    verbose=1  # Show detailed training progress (1 for progress bar, 0 for silent, 2 for per-epoch output)
)
```

```python
network.summary()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

history_dict = model.history

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.lineplot(x=range(1, len(history_dict['accuracy']) + 1), y=history_dict['accuracy'], label='Training Accuracy')
sns.lineplot(x=range(1, len(history_dict['val_accuracy']) + 1), y=history_dict['val_accuracy'], label='Validation Accuracy')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
sns.lineplot(x=range(1, len(history_dict['loss']) + 1), y=history_dict['loss'], label='Training Loss')
sns.lineplot(x=range(1, len(history_dict['val_loss']) + 1), y=history_dict['val_loss'], label='Validation Loss')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

**Now With CNN**

```python
network = Sequential()

# 1st Convolutional Layer: Extracts 32 feature maps using 3x3 filters
# Input shape (32,32,3) -> 32x32 RGB images (for example, CIFAR-100 dataset)
network.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))

# 1st Pooling Layer: Reduces feature map size by 2x using max pooling
network.add(MaxPooling2D((2,2)))

# 2nd Convolutional Layer: Extracts 64 feature maps using 4x4 filters
network.add(Conv2D(64, (4,4), activation='relu'))

# 2nd Pooling Layer: Further reduces feature map size
network.add(MaxPooling2D((2,2)))

# 3rd Convolutional Layer: Extracts 32 feature maps using 4x4 filters
network.add(Conv2D(32, (4,4), activation='relu'))

# 3rd Pooling Layer: Further reduces feature map size
network.add(MaxPooling2D((2,2)))

# Flattening Layer: Converts 2D feature maps into 1D vector for Dense layers
network.add(Flatten())

# Fully Connected Layer 1: 1024 neurons with ReLU activation
network.add(Dense(units=1024, activation='relu'))

# Dropout Layer: Drops 30% of neurons to prevent overfitting
network.add(Dropout(0.3))

# Fully Connected Layer 2: 512 neurons with ReLU activation
network.add(Dense(units=512, activation='relu'))
network.add(Dropout(0.3))  # Drop 30% of neurons

# Fully Connected Layer 3: 256 neurons with ReLU activation
network.add(Dense(units=256, activation='relu'))
network.add(Dropout(0.3))

# Fully Connected Layer 4: 128 neurons with ReLU activation
network.add(Dense(units=128, activation='relu'))
network.add(Dropout(0.3))

# Output Layer: 100 neurons (for 100 classes), Softmax for multi-class classification
network.add(Dense(units=100, activation='softmax'))
```

```python
network.summary()
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitor validation loss to decide when to stop training
    patience=5,  # Stop training if validation loss doesn't improve for 5 consecutive epochs
    restore_best_weights=True  # Restore the model weights from the epoch with the best validation loss
)  # Prevents overfitting by stopping training early when the model stops improving
```

```python
network.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model=network.fit(X_train,y_train,epochs=100,validation_data=(X_test,y_test),callbacks=[early_stopping],batch_size=64)
```

```python
model.summary()
```
